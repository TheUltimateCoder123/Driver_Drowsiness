```python
import os
import tqdm as tq
import shutil

direc = r'G:\Driver Drowsiness\mrlEyes_2018_01'

for (root, dirs, files) in os.walk(direc):
    for i in tq.tqdm([f for f in files if f.endswith('.png')]):     
        if i.split('_')[4]=='0':
            shutil.copy(src=root+'/'+i, dst=r'G:\Driver Drowsiness\Converted_data\close')
        
        elif i.split('_')[4]=='1':
            shutil.copy(src=root+'/'+i, dst=r'G:\Driver Drowsiness\Converted_data\open')

```
